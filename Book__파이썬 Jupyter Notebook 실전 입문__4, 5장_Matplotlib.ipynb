#### 피겨와 서브플롯
    Matplotlib에서 그래프 작성하기 전 '피겨 오브젝트(피겨)'와 '서브 플롯'을 생성할 필요가 있음.
    - 피겨 : 서브플롯 작성 영역
    - 서브플롯 : 그래프 작성하기 위한 영역 (Figure.add_subplot(), pyplot.subplots())

#### add_subplot() 메서드로 서브플롯 배치하기


```python
# 피겨 생성
fig = plt.figure()

# 개별 코드로 서브플롯 작성하기
# 피겨 안에 서브플롯을 3개 배치. 
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
```

```python
fig = plt.figure()

# 서브플롯 작성
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

# 번호 기술
for i, ax in enumerate([ax1, ax2, ax3], start=1):
    txt = 'ax{0}\n(22{0})'.format(i)
    ax.text(0.2, 0.4, txt, fontsize=24)

plt.show()
```

#### 두 개의 축을 가진 그래프 그리기

    Matplotlib에서 
    x축 공유해 2개 y축 가진 그림 작성할 경우, Axes.twinx() 메서드 사용
    y축 공유해 2개 x축 가진 그림 작성할 경우, twiny()메서드 사용.

```python
base_url = 'https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/'

t4816_csv = os.path.join(base_url, '4816.csv')
df = pd.read_csv(t4816_csv, index_col=0, parse_dates=['Date'])

fig = plt.figure(figsize =(10, 4))
ax1 = fig.add_subplot(111)

ax1.plot(df.index, df['Close'], color='b', label='stock price')

# X축 공유해서 Y축을 2개 사용하는 설정
ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color ='g', label = 'volume', width=2)

# 축과 축레이블 설정
ax1.set_yticks([i * 2000 for i in range(5)])
ax1.set_ylabel('stock price') 
ax2.set_yticks([i * 50000 for i in range(5)])

ax2.set_ylabel('volume') 
ax1.set_xlabel('Year and Month')

# 그래프 타이틀 설정
ax1.set_title('Price and Volume of the stock')

# 범례 설정
ax1.legend(loc=1)
ax2.legend(loc=2)

plt.show()
```

#### 여러가지 그룹에 대한 막대그래프 작성하기

```python
import numpy as np

# 데이터 세트 작성
x = [1, 2]
y1, y2, y3 = [1, 2], [2, 4], [3, 6]

# 복수 그룹의 막대 그래프
fig = plt.figure()
ax = fig.add_subplot(111)

w = 0.2
ax.bar(x, y1, label='y1')
ax.bar(x, y2, label='y2')
ax.bar(x, y3, label='y3')

ax.legend()

plt.show()
```

```python
# 복수 그룹의 막대그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111)

w = 0.2
ax.bar(x, y1, width=w, label='y1')
ax.bar(np.array(x) + w, y2, width=w, label='y2')
ax.bar(np.array(x) + w*2, y3, width=w, label='y3')

ax.legend()

plt.show()
```

#### 누적 막대그래프

    y1, y2, y3 값을 누적할 경우, 큰 값을 베이스로 먼저 까는 형태로 진행
    
    1) y1+y2+y3 값을 우선 그래프에 표시한 뒤
    2) y1+y2 값을 1) 위에 표시
    3) y1 값을 2) 위에 마지막으로 표시

```python
# 데이터 세트 작성
x = np.arange(5)
np.random.seed(0)
y = np.random.rand(15).reshape((3, 5))
y1, y2, y3 = y

y1b = np.array(y1)
y2b = y1b + np.array(y2)
y3b = y2b + np.array(y3)

#누적 막대 그래프 그리기
fig = plt.figure(figsize=(10, 3))
ax = fig.add_subplot(111)

ax.bar(x, y3b, label='y3')
ax.bar(x, y2b, label='y2')
ax.bar(x, y1b, label='y1')

ax.legend()

plt.show()
```

```python
# bottom 옵션으로 누적 설정하기 (두 개 그룹까지만 누적 시)

figure = plt.figure(figsize=(10, 3))
ax = figure.add_subplot(111)

ax.bar(x, y3, bottom=y2b, label='y3')
ax.bar(x, y2, bottom=y1, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()

plt.show()
```

    Axes.hist() 메서드 사용.

```python
# 막대의 폭과 수를 변경하는 경우
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x, rwidth=0.9, bins=40)

plt.show()
```

```python
# 근사 곡선 추가

bins = 50 #막대 수
dfmin = np.min(df['rating']) # 데이터 최솟값
dfmax = np.max(df['rating']) # 데이터 최댓값

# 히스토그램 그리기
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(df['rating'], bins = bins, range = (dfmin, dfmax), normed = True, rwidth = 0.9)

# 평균과 표준편차
mu, sigma = df['rating'].mean(), df['rating'].std()

# X값
x = np.linspace(dfmin, dfmax, bins)

# 근사적 확률밀도함수를 사용해 Y값 생성
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)** 2 / (2 * sigma **2))

# 근사곡선 그리기
ax.plot(x, y)

plt.show()
```

```python
# 여러 그룹을 겹쳐서 그리기
types = df['type'].unique()
labels = types.tolist()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

b_num = np.arange(0, 10.5, 0.5)

for t in types:
    ax.hist(df.loc[df['type']==t, 'rating'],
           bins = b_num,
           rwidth=0.9,
           alpha=0.5,
           label=t)

ax.legend()
ax.set_xlabel('rating')
ax.set_ylabel('Count(rating)')

plt.show()
```

## 5-6 그래프의 축과 눈금 설정

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.arange(0, 15, 0.2)
x2 = np.arange(0, 20, 0.2)

y1 = np.sin(x1)
y2 = np.sin(x2) * 2
y3 = np.cos(x2) * 2
y4 = np.cos(x1)

fig = plt.figure(figsize=(3, 3))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x2, y3)
ax4.plot(x1, y4)

fig.subplots_adjust(wspace=0.35, hspace=0.35)

plt.show()
```

```python
# 축의 범위를 일치시킨 경우

fig = plt.figure(figsize=(3, 3))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x1, y1, label='y1')
ax2.plot(x2, y2, label='y2')
ax3.plot(x2, y3, label='y3')
ax4.plot(x1, y4, label='y4')

# 축 범위의 설정
axes = [ax1, ax2, ax3, ax4]
for ax in axes:
    ax.set_xlim(0, 20)
    ax.set_ylim(-2, 2)

fig.subplots_adjust(wspace=0.35, hspace=0.35, bottom=0.2)

plt.show()
```

```python
# 축 설정 맞추기

#데이터 세트 작성
x1 = np.arange(0, 15, 0.2)
x2 = np.arange(0, 20, 0.2)

y1=np.sin(x1)
y2=np.sin(x2) * 2
y3=np.cos(x2) * 2
y4=np.cos(x1)

#피겨 오브젝트 생성
fig, axes=plt.subplots(2, 2, figsize=(3, 3), sharex=True, sharey=True)

# 그리기
axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x2, y3)
axes[1, 1].plot(x1, y4)

plt.show()
```

## 5-7 범례와 타이틀

```python
# 범례 설정하기

import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0, 20, 0.2)
y1=np.sin(x)
y2=np.cos(x)

fig=plt.figure()
ax=fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# 범례 표시
ax.legend()

plt.show()
```

```python
# 범례 배치하기

fig=plt.figure()

ax=fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# 범례표시 위치 조정
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# 서브플롯 위치 조정
fig.subplots_adjust(left=0.1, bottom=0.2, right=0.75, top=0.8)

plt.show()
```

```python
# 서브플롯 타이틀 설정하기

fig = plt.figure()

ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.set_title('sin and cos')

plt.show()
```

```python
# 피겨 타이틀 설정하기

fig=plt.figure()
ax=fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.set_title('sin and cos')

fig.suptitle('삼각함수')
plt.show()
```

```python
# 범례와 타이틀 서식 설정하기

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

fdict = {'family': 'serif',
        'style': 'italic',
        'weight': 'bold',
        'size': 12}
bdict = {'facecolor': 'k',
        'edgecolor': 'None',
        'alpha': 0.5,
        'boxstyle': 'larrow', #'l'eft 'arrow'b
        'linewidth': 2}

ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0, prop=fdict)

ax.set_title('sin and cos', fontdict=fdict, bbox=bdict, color='white')

fig.suptitle('삼각함수', y=0.95, size=16, color='grey', fontweight='bold')

fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8)

plt.show()
```
