### 1-4 한글 폰트 환경의 준비 
    Matplotlib에서 한글 처리할 때 종종 글자가 깨니는 문제가 있음. Matplotlib의 초기 설정에서 사용하는 폰트가 한글 설정에서 포함되지 않기 때문. 미리 폰트를 설정해줌으로서 해결.

```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(0)

x = range(5)
y = 10 + 5 * np.random.randn(5)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_title('한글을 지정한 타이틀')

ax.bar(x,y )
plt.show()
```

#### 1-4-1 한글 폰트 개별 설정하기

```python
import os
import numpy as np
from matplotlib import pyplot as plt, font_manager

# 폰트 캐쉬를 재구축
font_manager._rebuild()

if os.name == 'nt':
    # OS가 Window일 경우
    font_dir = font_manager.win32FontDirectory()
else:
    # OS가 macOS일 경우
    font_dir = '/Users/hugo/Library/Fonts/'
    
font_path = os.path.join(font_dir, 'NotoSansKR-Medium.otf')
font = font_manager.FontProperties(fname=font_path, size=14)

#그래프 작성
np.random.seed(0)
x = range(5)

y = 10 + 5 * np.random.randn(5)
fig = plt.figure()
ax = fig.add_subplot(111)

#여기서 fontproperties를 지정한다.
ax.set_title('한국어를 지정한 타이틀', fontproperties=font)
ax.bar(x,y)

plt.show()
```

#### 1-4-2 폰트 일괄 설정

```python
#rcParams을 활용해 폰트 일괄로 설정 -> 2021년 6월 기준 해당 정보 불필요하다 판단하여 해결 중단

import os
import numpy as np
from matplotlib import pyplot as plt, rcParams

#여기서 지정된 설정은 초기 설정 값으로 이용
rcParams['font.sans-serif'] = 'NotoSansKR-Medium.otf'
rcParams['font.weight'] = 'medium'
rcParams['axes.titlesize'] = 15
rcParams['ytick.labelsize'] = 12
rcParams['xtick.labelsize'] = 12

np.random.seed(0)

x = range(5)
y = 10 + 5 * np.random.randn(5)
fig = plt.figure()
ax = fig.add_subplot(111)

# 여기에서 폰트 정보를 지정하지 않아도 됨.
ax.set_title('한국어를 지정한 타이틀')
ax.bar(x, y)

plt.show()
```

#### Docstring 표시
    정의 : Python의 모듈이나 클래스, 함수의 코멘트.
    방법 : 1. 그래프 명이나 함수명 끝에 '?'를 붙이면 Docstring을 참조할 수 있다.   
    
#### Tool tip 표시
    방법 : Shift + Tab
    
#### 자동완성
    방법 : 일부 코드 작성 후 'Tab'키 누르기.
    예 : math.s + Tab키

```python
import math
math.sqrt(4)
```
